### **DDA Algo practical 4A**

```python
#dda code  same like algo
def dda(x1,y1,x2,y2):
  xdelta=x2-x1
  ydelta=y2-y1
  xcord=[]
  ycord=[]# creating list to save the final cordinates
  if(xdelta>=ydelta):
    length=xdelta
  else:
    length=ydelta
  dx=(x2-x1)/length
  dy=(y2-y1)/length
  x=x1+0.5*np.sign(dx)
  y=y1+0.5*np.sign(dy)
  i=1
  while(i<=length):
    xcord.append(x)#appending the cordinate into the list
    ycord.append(y)
    x=x+dx
    y=y+dy
    i=i+1
  return (xcord,ycord)#returning the list
x,y=dda(0,0,6,6)#calling dda function
plt.plot(x,y,"y.-")#ploting the points in graph
plt.xlabel("X-Axis")#title of the x axis
plt.ylabel("Y-Axis")#title of the y axis
plt.title("DDA Algorithm")#for title of the graph
plt.show()#displaying the figure

```

### **Bresenham**

```python
def bresenham(x1,y1,x2,y2):
  xdelta=x2-x1
  ydelta=y2-y1
  x=x1
  y=y1
  e=2*ydelta-xdelta
  xcords=[]
  ycords=[]
  i=1

  while(i<=xdelta):
    xcords.append(x)
    ycords.append(y)

    while(e>=0):
      y+=1
      e=e-2*xdelta
    x+=1
    e=e+2*ydelta


    i=i+1

  return (xcords,ycords)

x,y=bresenham(0,0,6,6)
#y.- means
# y = yellow color
#. -  dots in line
# - means straight line
#markersize size of dot
plt.plot(x,y,"y.-",label="Bresenham Algorithm",markersize=20,)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Bresenham Algorithm")
plt.legend()
plt.show()

```



### **Mid Point Circle Algo Practical 5**


```python
import matplotlib.pyplot as plt

def midpoint(r):
    x = 0
    y = r
    d = 1.25 - r
    xcords = []
    ycords = []

    while x <= y:

        xcords.append([x, -x, x, -x, y, -y, y, -y])
        ycords.append([y, y, -y, -y, x, x, -x, -x])

        if d < 0:
            x += 1
            d += 2 * x + 1
        else:
            x += 1
            y -= 1
            d += 2 * x + 2 * y + 1

    return (xcords, ycords)

# Calling  function with radius 6
x, y = midpoint(6)


plt.plot(x, y, "y.")
plt.axis("equal")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Midpoint Circle Algorithm")

plt.grid()
plt.show()
```

### **Practical 6A Translation only**






```python
def translation(x,y,tx,ty):
  x1=x+tx#adding original cordinate with translation factor to get new cordinate
  y1=y+ty
  return (x1,y1)


x=10
y=20
plt.scatter(x,y,label="Original Cordinate")
x1,y1=translation(x,y,5,10)
plt.scatter(x1,y1,label="Translated Cordinate")

plt.plot([x,x1],[y,y1],"b--",label="Translation")
plt.legend()


plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Translation ")
plt.legend()

plt.show()
```
