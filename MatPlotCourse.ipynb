### **DDA Algo practical 4A**

```python
#dda code  same like algo
def dda(x1,y1,x2,y2):
  xdelta=x2-x1
  ydelta=y2-y1
  xcord=[]
  ycord=[]# creating list to save the final cordinates
  if(xdelta>=ydelta):
    length=xdelta
  else:
    length=ydelta
  dx=(x2-x1)/length
  dy=(y2-y1)/length
  x=x1+0.5*np.sign(dx)
  y=y1+0.5*np.sign(dy)
  i=1
  while(i<=length):
    xcord.append(x)#appending the cordinate into the list
    ycord.append(y)
    x=x+dx
    y=y+dy
    i=i+1
  return (xcord,ycord)#returning the list
x,y=dda(0,0,6,6)#calling dda function
plt.plot(x,y,"y.-")#ploting the points in graph
plt.xlabel("X-Axis")#title of the x axis
plt.ylabel("Y-Axis")#title of the y axis
plt.title("DDA Algorithm")#for title of the graph
plt.show()#displaying the figure

```

### **Bresenham**

```python
def bresenham(x1,y1,x2,y2):
  xdelta=x2-x1
  ydelta=y2-y1
  x=x1
  y=y1
  e=2*ydelta-xdelta
  xcords=[]
  ycords=[]
  i=1

  while(i<=xdelta):
    xcords.append(x)
    ycords.append(y)

    while(e>=0):
      y+=1
      e=e-2*xdelta
    x+=1
    e=e+2*ydelta


    i=i+1

  return (xcords,ycords)

x,y=bresenham(0,0,6,6)
#y.- means
# y = yellow color
#. -  dots in line
# - means straight line
#markersize size of dot
plt.plot(x,y,"y.-",label="Bresenham Algorithm",markersize=20,)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Bresenham Algorithm")
plt.legend()
plt.show()

```



### **Mid Point Circle Algo Practical 5**


```python
import matplotlib.pyplot as plt

def midpoint(r):
    x = 0
    y = r
    d = 1.25 - r
    xcords = []
    ycords = []

    while x <= y:

        xcords.append([x, -x, x, -x, y, -y, y, -y])
        ycords.append([y, y, -y, -y, x, x, -x, -x])

        if d < 0:
            x += 1
            d += 2 * x + 1
        else:
            x += 1
            y -= 1
            d += 2 * x + 2 * y + 1

    return (xcords, ycords)

# Calling  function with radius 6
x, y = midpoint(6)


plt.plot(x, y, "y.")
plt.axis("equal")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Midpoint Circle Algorithm")

plt.grid()
plt.show()
```

### **Practical7a Rotation**

```python
def rotate(x,y,theta):
  theta=np.radians(theta)

  x=x
  y=y
  xcord=[]
  ycord=[]
  for i in range(len(x)):#len(y) bhi dal sak the hai
  #as per rotation matrix
    x1 = x[i] * np.cos(theta) - y[i] * np.sin(theta)#x'=xcos(0)-ysin(0)
    y1 = x[i] * np.sin(theta) + y[i] * np.cos(theta)#y'=xsin(0)-ycos(0)
    xcord.append(x1)
    ycord.append(y1)

  return xcord,ycord
x=[10,20,30]#original cordinate
y=[40,40,60]
theta=45
x1,y1=rotate(x,y,theta)
print(x1,y1)



plt.plot(x,y,'r.-',label="orignal ")
plt.plot(x1,y1,'y.--',label="rotated")
plt.legend()





plt.grid(True)
plt.title("Rotation")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()





```

```python
import numpy as np
import matplotlib.pyplot as plt

def scale_and_translate(x, y, sx, sy, tx, ty):
    scaled_x = x * sx
    scaled_y = y * sy
    translated_x = scaled_x + tx
    translated_y = scaled_y + ty
    return np.array([translated_x, translated_y])

house = np.array([
    [1, 1],
    [1, 2],
    [2, 2],
    [2, 1],
    [1, 1],  # Close the base
    [1, 2],  # Start the roof
    [1.5, 3],
    [2, 2]   # Close the roof
])

plt.plot(house[ :,0], house[:,1], color='black', label='Original House')
newhouse = scale_and_translate(house[ :,0], house[:,1] , 1.5, 1.5, 2, 2)
plt.plot(newhouse[0], newhouse[1], "r.--", label="Scaled and Translated House")
newhouse2=scale_and_translate(newhouse[0],newhouse[1],2,2,3,3)
plt.plot(newhouse2[0],newhouse2[1],"b.--")
plt.legend()
plt.grid(True)
plt.title("Scaling and Translation")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()

```
